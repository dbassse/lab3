**Задание 7**. Трехмерный график поверхности. Используя mpl_toolkits.mplot3d , постройте 3D-график функции $z = sin(√x2 + y2)$
на сетке значений $x$, $y$ в диапазоне [−5, 5].

```python
import mpl_toolkits.mplot3d 
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(
    x, y, z,
    cmap='viridis'
)
```

**Задание 12. Изменение уровня освещенности в течении дня.**<br>
Фиксировался уровень освещенности в люксах (lx) на улице в зависимости от времени суток:<br>
- Время суток (часы): [0, 3, 6, 9, 12, 15, 18, 21, 24]
- Освещенность (lx): [0, 5, 500, 15000, 40000, 35000, 10000, 100, 0]
<br>
Используйте заливку области под графиком и градиент цветов.

```python
import matplotlib.pyplot as plt
import numpy as np
time = [0, 3, 6, 9, 12, 15, 18, 21, 24]
lux = [0, 5, 500, 15000, 40000, 35000, 10000, 100, 0]
fig, ax = plt.subplots(figsize=(12, 6))

#Нормализую данные
cmap = plt.get_cmap('viridis')
norm = plt.Normalize(min(lux), max(lux))

#Подбираем для каждой пары точек подходящее место на цветовой карте
for i in range(len(time) - 1):
    ax.fill_between(
        [time[i], time[i+1]],
        [lux[i], lux[i+1]],
        color=cmap(norm(lux[i])),
    )
# Сам график)
ax.plot(time, lux)
plt.title("Освещенность")
plt.xlabel('Время суток')
plt.ylabel('Количество света')
ax.grid(True)
```

##**Задачи на вычисление определенного интеграла с помощью Matplotlib**<br>
В каждой задаче требуется:
- **Построить график** подинтегральной функции.
- **Вычислить площадь под кривой** на заданном отрезке как значение
определенного интеграла.
- **Закрасить область под графиком**, чтобы визуализировать интеграл.<br>
Вычислите площадь под кривой:<br>
$f(x)=|sin(x)|$<br>
На отрезке: $[0;2\pi]$

```python
def f(x): return np.abs(np.sin(x))
x = np.linspace(0, 2*np.pi, 500)
y = f(x)

# Разбиваем площадь под графиком на трапеции и находим их площадь с помощью встроенной функции
dx = x[1] - x[0]
integral = np.trapz(y, x, dx)

plt.figure(figsize=(8, 5))
plt.plot(x, y)

plt.fill_between(x, y)

plt.title('$\\int_{0}^{2\\pi} |\\sin(x)| \\, dx$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

##**Задачи на построение 3D-графиков с помощью Matplotlib**<br>
Во всех задачах требуется:
- **Построить трехмерный график** функции $f(x, y)$ в заданных пределах.
- **Использовать библиотеку Matplotlib** для визуализации.
- **Оформить график:** добавить заголовок, подписи осей и цветовую карту (если
уместно).<br>
Постройте поверхность:<br>
$f(x,y)=e^{-((x-1)^2 + y^2)}+ e^{-((x+1)^2 + y^2)}$<br><br>
На $x,y ∈ [-3;3]$


```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)

Z = np.exp(-((X - 1)**2 + Y**2)) + np.exp(-((X + 1)**2 + Y**2))

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='inferno')

# Настройки графика
ax.set_title("$e^{-((x-1)^2 + y^2)} + e^{-((x+1)^2 + y^2)}$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

# Цветовая шкала
fig.colorbar(surf)
plt.show()
```
